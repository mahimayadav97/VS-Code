```python
from groundingdino.util.inference import load_model, load_image, predict, annotate
import os
from tqdm import tqdm
import yaml
from PIL import Image
import glob

# Load the Grounding DINO model 
dino = load_model('GroundingDINO/groundingdino/config/GroundingDINO_SwinT_OGC.py', 
                  'D:/Dissertation/Validation of pretrained model/GroundingDINO/weights/groundingdino_swint_ogc.pth')

def annotate_new_dataset(dino, dataset_path, output_dir, data_size=3000):
    # Create output directories for images and labels
    image_dir = f'{output_dir}/images'
    label_dir = f'{output_dir}/labels'
    os.makedirs(image_dir, exist_ok=True)
    os.makedirs(label_dir, exist_ok=True)

    # Load images from the dataset path
    image_paths = glob.glob(f'{dataset_path}/*.jpg')  
    image_paths = image_paths[:data_size]  # Limit to data_size
    
    for i, image_path in enumerate(tqdm(image_paths)):
        image = Image.open(image_path).resize((640, 640))
        image_output_path = f'{image_dir}/{i:06d}.png'
        label_output_path = f'{label_dir}/{i:06d}.txt'
        image.save(image_output_path)
        
        image_source, image = load_image(image_output_path)
        boxes, logits, phrases = run_dino(dino, image, text_prompt='leaf')  
        label = ['0 ' + ' '.join(list(map(str, b))) for b in boxes.tolist()]
        label = '\n'.join(label)
        with open(label_output_path, 'w') as f:
            f.write(label)

# Directory paths
dataset_path = 'D:/Dissertation/Validation of pretrained model/Leaf GroundingDINO Dataset'
output_dir = 'data/leaf'

# Annotate the new dataset
annotate_new_dataset(dino, dataset_path, output_dir, data_size=3000)

# Create and save the YOLO data configuration file
config = {
    'names': ['leaf'], 
    'nc': 1,
    'train': 'leaf/images',
    'val': 'leaf/images'  
}

with open(f'{output_dir}/data.yaml', 'w') as f:
    yaml.dump(config, f)
```
